Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the California House Price dataset

```python
house_price_dataset = fetch_california_housing()
print(house_price_dataset)
```

Transforming our California dataset into a Dataframe for analysis

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

Printing the dataframe

```python
print(house_price_dataframe.head())
print(house_price_dataframe.columns)
print(house_price_dataset.target)
print(house_price_dataset.target_names)
```

Add the target column (Median House Value) to the dataframe

```python
house_price_dataframe['MedHouseVal'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

Checking the number of rows and columns in the dataframe

```python
house_price_dataframe.shape
```

Check for missing values (If we had missing values, we would have to implement methods like dropping or imputation)

```python
house_price_dataframe.isnull().sum()
```

Statistical measures of the dataset

```python
house_price_dataframe.describe()
```

Understanding the correlation between various features of the dataset

```python
correlation = house_price_dataframe.corr()
```

Constructing a heatmap to understand the correlation

```python
plt.figure(figsize = (10,10))
#color bar, square view, one value after the decimal point, annotations, annotation size, color of the heatmap
sns.heatmap(correlation, cbar = True, square = True , fmt = '.1f', annot=True, annot_kws={'size': 8}, cmap= 'Blues')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['MedHouseVal'], axis = 1) #dropping the whole column
y = house_price_dataframe['MedHouseVal']

print(X)
print(y)
```

Split the data into training data and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)  #using same value guarantees splitting the data in the same way/reproduce the code
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training (Using the XGBoost Regressor Model, a type of decision tree algorithm)

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, y_train)
```

Prediction and Evaluation

```python
#we cannot use accuracy for regression problems
#instead, we use R^2 error and Mean Squared Error

#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction) #y_train are the original values
```

```python
#1. Use R^2 Error
score_1 = metrics.r2_score(y_train, training_data_prediction) #finds the variance the values in both

#2. Mean Absolue Error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)

print("R^2 Error: ", score_1)
print("Mean Absolute Error: ", score_2)
```

Visualize Actual Prices v.s. Predicted Prices

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices v.s. Predicted Prices")
plt.show()
```

We also need to find the accuracy and the error values for the testing data as well

```python
testing_data_prediction = model.predict(X_test)

score_3 = metrics.r2_score(y_test, testing_data_prediction)
score_4 = metrics.mean_absolute_error(y_test, testing_data_prediction)

print("R^2 Error: ", score_3)
print("Mean Absolute Error: ", score_4)
```
