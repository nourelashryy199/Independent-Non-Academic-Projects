```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Separating the features and the target

```python
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

```python
print(X,Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
print(X_test_std)
```

Building the Neural Network

```python
# importing tensorflow and keras
import tensorflow as tf #deep learning library developed by google
tf.random.set_seed(3) #get the same accuracy score no matter how many times you run your code
from tensorflow import keras #you can set your backend to tensorflow or pytorch in keras
```

```python
#setting up the layers of nn
model = keras.Sequential([

                          keras.layers.Flatten(input_shape = (30,)),
                          keras.layers.Dense(20, activation = 'relu'),
                          keras.layers.Dense(40, activation = 'relu'),
                          keras.layers.Dense(60, activation = 'sigmoid'),
                          keras.layers.Dense(2, activation = 'sigmoid')
]) #stacking layers of nn
#flatten converts our X_train to a single dimensional array such that for each feature, we have one input neuron
#i.e. number of neurons in input layer = number of features that you have
#number of neurons in the last layer is equal to the number of classes that we are working with i.e. benign or malignant

```

```python
#compiling the nn
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# training the nn
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 20)
#validation split is for validating on 10% of the training data
```

Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model (Neural Network) Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc = 'lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model (Neural Network) Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc = 'upper right')
plt.show()
```

Get accuracy value of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test) #part of keras
print(accuracy)
```

model.predict() gives the prediction probability of each class for a specific data point

```python
#argmax function
my_list = [0.25, 1.75, 1.00]
index_of_max_value = np.argmax(my_list)
print(index_of_max_value)
```

```python
#converting the prediction probability to class labels
Y_pred_labels = [int(np.argmax(i)) for i in Y_pred] #for loop
print(Y_pred_labels)
```

Building the Predictive System

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

#change the input_data to numpy array

input_data_as_np_array = np.asarray(input_data)

#reshape numpy array as we are predicting for one data point
input_data_reshape = input_data_as_np_array.reshape(1,-1)

#standardize the input data
input_data_std = scaler.transform(input_data_reshape)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [int(np.argmax(prediction))]
print(prediction_label)

if (prediction_label[0] == 0):
  print('The tumor is malignant')
else:
  print('The tumor is benign')
```
